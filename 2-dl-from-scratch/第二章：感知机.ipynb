# 实现逻辑电路

先实现一个 `AND` 函数
与门

```python
def AND(x1, x2):
    w1, w2, theta = 0.5, 0.5, 0.7
    if w1 * x1 + w2 * x2 <= theta:
        return 0
    else:
        return 1
```

```python
AND(0, 0)
AND(1, 1)
```

使用 `numpy` 实现 `AND` --- 与门

```python
import numpy as np
def AND(x1, x2):
    w = np.array([0.5, 0.5])
    x = np.array([x1, x2])
    b = -0.7  # b是被称为偏置的参数，用于控制神经元被激活的容易程度
    #     if b + w * x <= 0:
    '''
    你用上述的 if 不对，因为 w * x 之后还是一个数组！
    '''
    if b + sum(w * x) <= 0:
        return 0
    else:
        return 1
AND(1, 0)
```

实现与非门

```python
def NAND(x1, x2):
    x = np.array([x1, x2])
    w = np.array([-0.5, -0.5])
    b = 0.7
    tmp = b + sum(x * w)
    if tmp <= 0:
        return 0
    else:
        return 1
NAND(1, 1)
```

实现或门

```python
def OR(x1, x2):
    x = np.array([x1, x2])
    w = np.array([0.5, 0.5])
    b = -0.2
    tmp = b + sum(x * w)
    if tmp <= 0:
        return  0
    else:
        return 1
```

# 总结
实现 与门、 或门、与非门 之间的区别就是 对于 权重、和偏置设置的不一样 类实现这三个门电路之间的区别\
在这里主要就是体会 权重 和 偏置 设置的不同来实现的！

## 实现异或门

```python
def XOR(x1, x2):
    a = OR(x1, x2)
    b = NAND(x1, x2)
    return AND(a, b)
```

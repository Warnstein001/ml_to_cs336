#### Sigmoid Function

```python
import numpy as np
```

```python
def sigmoid_function(x):
    return 1 /(1 + np.exp(-x))
```

```python
sigmoid_function(1)
```

```python
import matplotlib.pylab as plt
```

```python
# 生成数据
x = np.linspace(-10, 10, 100)
y = sigmoid_function(x)
# 画出图表
plt.figure(figsize=(8, 4)) # 设置画布大小
plt.plot(x, y) # 绘制Sigmoid曲线
plt.title("Sigmoid Function Curve") # 标题
plt.xlabel('x') # x轴标签
plt.ylabel('sigmoid(x)') # y轴标签
plt.show() # 显示图像
```

#### Step Function

```python
def step_function(x):
    # 法一
    # return np.where(x > 0, 1, 0) # 这个是完成三目运算符的功能
    #----
    # 法二
    y = x > 0
    return y.astype(np.int_)
```

```python
# 生成数据
x = np.linspace(-10, 10, 100)
y = step_function(x)
# 画出图表
plt.figure(figsize=(8, 4))
plt.plot(x, y)
plt.title("Step Function Curve")
plt.xlabel("x")
plt.ylabel("step(x)")
plt.show()
```

#### ReLU Function

```python
def ReLU(x):
    # 法一
    # return np.where(x > 0, x, 0)
    # 法二
    return np.maximum(0, x)
```

```python
# 生成数据
x = np.linspace(-6, 6, 100)
y = ReLU(x)
# 画出图像
plt.figure(figsize=(8, 4))
plt.plot(x, y)
plt.title("ReLu Function Curve")
plt.xlabel("x")
plt.ylabel("ReLU(x)")
plt.ylim(-1, 5)
plt.show()
```

#### Softmax Function

```python
def softmax_function(x):
    exp_a = np.exp(x)
    sum_exp_a = np.sum(exp_a)
    y = exp_a / sum_exp_a
    return y
```

```python
a = np.array([1010, 1000, 990])
softmax_function(a)
```

上面我们看到，他输出的是 nan （not a number，不确定） 其就是溢出了，因为 $e^x$ 可是程爆炸及增长的，那就可以通过离散化的方法防止溢出

```python
def softmax_function(x):
    c = np.max(x)
    exp_a = np.exp(x - c) # 离散化 -- 防溢出
    sum_exp_a = np.sum(exp_a)
    y = exp_a / sum_exp_a
    return y
```

```python
a = np.array([1010, 1000, 990])
softmax_function(a)
```
